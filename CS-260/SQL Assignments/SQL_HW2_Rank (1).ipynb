# QUESTION 1

When performing data analysis, you should be very aware of which columns have missing values.  Doing so will help you be careful about using COUNT(\*) or COUNT(col_name) in appropriate places.

Generate a table showing how many nulls there are per column.  

Rename the columns as shown.

So your final answer should match the table below, exactly.

| missing_donation | missing_state | missing_affiliation | missing_class_year | missing_preferred_division | missing_date_submitted | missing_donation_year |
|:-----------------:|:--------------:|:-------------------:|:------------------:|:--------------------------:|:----------------------:|:---------------------:|
|         0         |        3       |          0          |         683        |             936            |            0           |           0           |
|                   |                |                     |                    |                            |                        |                       |

````python
# CodeGrade Tag: question_1'

sql_statement1 = """SELECT COUNT(*)-COUNT(donation) AS missing_donation, COUNT(*)-COUNT(state) AS missing_state, COUNT(*)-COUNT(affiliation) AS missing_affiliation, COUNT(*)-COUNT(class_year) AS missing_class_year, COUNT(*)-COUNT(preferred_division) AS missing_preferred_division, COUNT(*)-COUNT(date_submitted) AS missing_date_submitted, COUNT(*)-COUNT(donation_year) AS missing_donation_year
                    FROM hearts_on_fire;"""

query_results1 = pd.read_sql_query(sql_statement1, conn)
query_results1
````

# QUESTION 5

Recall that the Northeast Region is CT, ME, MA, NH, RI, VT, NJ, NY, PA.
  
Write a query to determine which states OUTSIDE of the Northeast Region had the 3 largest amounts of alumni contributors in the year 2021.

Your columns should be pertinent to the question at hand and your final table should only have 3 rows, making your answer clear.


````python
# CodeGrade Tag: question_5
sql_statement5 = """SELECT state, COUNT(affiliation) AS num_contributors, donation_year
                    FROM hearts_on_fire
                    WHERE (state NOT IN ('CT', 'ME', 'MA', 'NH', 'RI', 'VT', 'NJ', 'NY', 'PA')) AND (donation_year = 2021) AND (affiliation = 'Alumni')
                    GROUP BY state
                    ORDER BY COUNT(state) DESC
                    LIMIT 3;"""

query_results5 = pd.read_sql_query(sql_statement5, conn)
query_results5
````

# QUESTION 6

For the year 2020, show the number of donors per the amount of donation.

````python
# CodeGrade Tag: question_6
sql_statement6 = """SELECT donation, COUNT(donation) AS num_donors
                    FROM hearts_on_fire
                    WHERE (donation_year = 2020)
                    GROUP BY donation
                    ORDER BY donation ASC;"""

query_results6 = pd.read_sql_query(sql_statement6, conn)
query_results6
````

# QUESTION 7

There were a lot of 1s in the prior table.  

Run the same query but this time, show only those rows where more than 1 person contributed the given amount.

````python
# CodeGrade Tag: question_7
sql_statement7 = """SELECT donation, COUNT(donation) AS num_donors
                    FROM hearts_on_fire
                    WHERE (donation_year = 2020)
                    GROUP BY donation
                    HAVING COUNT(donation) > 1
                    ORDER BY donation ASC;"""

query_results7 = pd.read_sql_query(sql_statement7, conn)
query_results7
````

# QUESTION 11

# QUESTION 12

Ask an interesting question that was not already asked, one that uses GROUPBY.  

Make it interesting and informative, ie, don't give me a lame question to just get points.

What is the total amount of donations from students in the class of 2021 within the Northeast Region per state, ordered from the lowest total to the highest.

````python
# CodeGrade Tag: question_12

sql_statement12 = """SELECT state, SUM(donation) AS total_donations
                     FROM hearts_on_fire
                     WHERE (state IN ('CT', 'ME', 'MA', 'NH', 'RI', 'VT', 'NJ', 'NY', 'PA')) AND (affiliation = 'Student') AND (class_year = 2021)
                     GROUP BY state
                     ORDER BY total_donations ASC;"""

query_results12 = pd.read_sql_query(sql_statement12, conn)
query_results12
````

# QUESTION 13

The following query fails in DataCamp.  

``` SELECT country, AVG(gross) ```

```FROM films```

```WHERE AVG(gross) > 1000000```

```GROUP BY country```


### Why?-  What's Wrong?

The query will fail in DataCamp due to the fact that it is trying to run a query that contains a WHERE function which contains an aggregate function. When using WHERE, you cannot use aggregate functions. To fix the error, we could instead use the HAVING function after GROUP BY. This would be HAVING AVG(gross) > 1000000. The HAVING function is what we must use whenever we wish to utilize an aggregate function within a WHERE search.
